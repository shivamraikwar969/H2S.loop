```python
# 1. Install Ultralytics
!pip install ultralytics
from ultralytics import YOLO
import os
```

```python
# 2. Prepare Dataset
# Ensure dataset.zip is uploaded in /content/
if os.path.exists('/content/dataset.zip'):
    !unzip -q /content/dataset.zip -d /content/
    print("✅ Dataset extracted successfully.")
else:
    print("❌ dataset.zip not found! Please upload the file.")
```

```python
# 3. Verify data.yaml
# YOLOv8 needs the absolute path in data.yaml. We'll update it automatically.
yaml_path = '/content/datasets/fire_smoke/data.yaml'
if os.path.exists(yaml_path):
    with open(yaml_path, 'r') as f:
        content = f.read()
    
    # Replace local path with Colab path if needed
    if 'path:' in content:
        import re
        content = re.sub(r'path: .*', 'path: /content/datasets/fire_smoke', content)
    
    with open(yaml_path, 'w') as f:
        f.write(content)
    
    print("✅ data.yaml updated for Colab paths.")
    print("---")
    !cat {yaml_path}
else:
    print("❌ data.yaml missing!")
```

```python
# 4. Training
# Use YOLOv8n (nano) for speed and efficiency.
model = YOLO('yolov8n.pt')

model.train(
    data=yaml_path,
    epochs=50,
    imgsz=640,
    batch=16,
    name='fire_smoke_v8',
    augment=True,
    plots=True
)
```

```python
# 5. Export Model weights
import shutil
from google.colab import files

best_weights = '/content/runs/detect/fire_smoke_v8/weights/best.pt'
target_name = 'fire_smoke_v8.pt'

if os.path.exists(best_weights):
    shutil.copy(best_weights, target_name)
    print(f"✅ Model trained and renamed to {target_name}")
    print("📥 Downloading model...")
    files.download(target_name)
else:
    print("❌ Weights not found! Check training logs.")
```
